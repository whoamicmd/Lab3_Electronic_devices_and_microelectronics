За статистичними характеристиками для схем **(СЕ)** та **(СБ)** знайдемо h параметри для обох схем. При розрахунку параметрів використовуємо вихідну характеристику середнього та найближчого до середнього значення вхідного струму. Отримані значення параметрів h запишемо до таблиці 3.1.

Розрахунок значень h параметрів.
Вхідний опір:
$$
h_{11e} = \frac{\Delta U_{БЕ}}{\Delta I_Б}
$$
Коефіцієнт зворотного зв'язку по напрузі:
$$
h_{12e} = \frac{\Delta U_{БЕ}}{\Delta U_{КЕ}}
$$
Коефіцієнт передачі струму:
$$
h_{21e} = \frac{\Delta I_К}{\Delta I_{Б}}
$$
Вихідна провідність:
$$
h_{22e} = \frac{\Delta I_К}{\Delta U_{КЕ}}
$$

```python
# Дані для СЕ та СБ у вигляді словників
data_se = {
    "ube": [772e-3, 822e-3, 828e-3],
    "uke": [1, 3, 10, 20],
    "ib":  [0.8e-3, 1.2e-3, 750e-6, 1.75e-3],
    "ik":  [73.65e-3, 76.31e-3, 53.58e-3]
}

data_sb = {
    "ube": [734e-3, 795e-3, 803e-3],
    "uke": [5, 25, 10, 0],
    "ib":  [40e-3, 60e-3, 25e-3, 85e-3],
    "ik":  [59.15e-3, 59.38e-3, 39.6e-3]
}

# Функція для розрахунку h-параметрів
def calc_h(data):
    ube, uke, ib, ik = data["ube"], data["uke"], data["ib"], data["ik"]
    h11 = (ube[1] - ube[0]) / (ib[3] - ib[2])
    h12 = (ube[2] - ube[1]) / (uke[3] - uke[2])
    h21 = (ik[1] - ik[2]) / (ib[1] - ib[0])
    h22 = (ik[1] - ik[0]) / (uke[1] - uke[0])
    return h11, h12, h21, h22

# Функція виводу
def show_result(name, hes, scale=1e3, unit="*10^-3"):
    print(f"=== Розрахункові параметри схеми ({name}) ===")
    print(f"h11e (вхідний опір)          = {hes[0]:.1f} Ом")
    print(f"h12e (зворотний зв'язок U)   = {hes[1]*scale:.2f} {unit}")
    print(f"h21e (β, підсилення по струму) = {hes[2]:.1f}")
    print(f"h22e (вихідна провідність)   = {hes[3]:.5f} См  →  {hes[3]*scale:.1f} мСм\n")
    
# Розрахунок
h_se = calc_h(data_se)
h_sb = calc_h(data_sb)
h_se_dv = (2200, 3.6e-4, 70, 0.0125e-3)

# Вивід результатів
show_result("СЕ", h_se, 1e3, "*10^-3")
show_result("СБ", h_sb, 1e4, "*10^-4")
```

Розрахуємо для схем СЕ та СБ значення фізичних параметрів $r_е$, $r_б$, $r_к$, $µ_{ек}$, а також коефіцієнтів α та β. Запишемо отримані дані у таблицю 3.2. 

$$
r_{e} = \frac{h_{12e}}{h_{22e}};
$$
$$
r_{б} = h_{11e} - (1 + h_{21e})\cdot \frac{h_{12e}}{h_{22e}};
$$
$$
r_x = \frac{1 + h_{21e}}{h_{22e}};
$$
$$
\mu_{вк} = \frac{h_{11e} \cdot h_{22e}}{(1 + h_{21e}) - h_{12e}};
$$
$$
\alpha = \frac{h_{21e}}{1 + h_{21e}};
$$
$$
\beta = h_{21e}.
$$

```python
# Функція для розрахунку
def calc_r(data):
    h11, h12, h21, h22 = data[0], data[1], data[2], data[3]
    re = h12 / h22
    rb = h11 - (1 + h21) * re
    rx = (1 + h21) / h22
    mu = (h11 * h22) / ((1 + h21) - h12) 
    a = h21 / (1 + h21)          # коефіцієнт передачі струму α ≈ β/(β+1)
    b = h21
    return re, rb, rx, mu, a, b

def show_results(name, res, scale=1e3, unit="*10^-3"):
    print(f"=== Розрахункові параметри схеми ({name}) ===")
    print(f"re = {res[0]:.1f} Ом")
    print(f"rb = {res[1]:.1f} Ом")
    print(f"rx = {res[2]:.0f} Ом")
    print(f"μ  = {res[3]*scale:.5f} {unit}")
    print(f"α  = {res[4]:.4f}")
    print(f"β  = {res[5]:.1f}\n")

# Розрахунок
r_se = calc_r(h_se)
r_sb = calc_r(h_sb)
r_se_dv = calc_r(h_se_dv)
                                          
# Вивід
show_results("СЕ", r_se, 1e3, "*10^-3")
show_results("СБ", r_sb, 1e6, "*10^-6")
show_results("СЕ Довідникові значення", r_se_dv, 1e6, "*10^-6")
```
